```python
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

!pip install xlrd
```

```python
df = pd.read_csv('lynxhare.csv', delimiter=';')
df = df[(df['hare'].notna()) & (df['lynx'].notna())]
df = df[['year', 'hare', 'lynx']]
df = df.set_index('year')
```

```python
def normalize_column(series):
    return series / np.mean(series)
df['hare'] = normalize_column(df['hare'])
df['lynx'] = normalize_column(df['lynx'])
```

```python
ax = df.plot();
ax.set_ylabel('Normalized Population Size');
ax.set_title('Lynx Hare Populations');
```
